```python
df.drop(['company', 'agent', 'reservation_status', 'reservation_status_date'], axis=1, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Filling the missing 'children' with 0
df['children'].fillna(0, inplace=True)

# Filling the missing 'country' with the most frequent value
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.info()
```

```python
# Creating a proper date column
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' +
                                    df['arrival_date_month'] + '-' +
                                    df['arrival_date_day_of_month'].astype(str),
                                    format='%Y-%B-%d')
```

```python
df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```
